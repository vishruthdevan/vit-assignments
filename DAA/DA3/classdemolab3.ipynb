```python
import math


def knn(points, p, k=3):
    distances = []
    frequencies = {}
    for group in points:
        frequencies[group] = 0
        for point in points[group]:
            euclidean_distance = 0
            for i, j in zip(point, p):
                euclidean_distance += (i-j)**2
            euclidean_distance = math.sqrt(euclidean_distance)
            distances.append((euclidean_distance, group))
    distances = sorted(distances)[:k]
    for i in distances:
        frequencies[i[1]] += 1
        
    max = list(frequencies.keys())[0]
    for i in frequencies:
        if frequencies[i] > frequencies[max]:
            max = i
            
    return max
```

```python
X_train_list = X_train.values.tolist()
y_train_list = y_train.values.tolist()
#converting dataframe values to a list

divided_by_category = {x:[] for x in y_train.unique()}
# divided_by_category is a dictionary of the form {'B':[points with category 'B'], 'M':[points with category 'M']}

for i, j in zip(X_train_list, y_train_list):
    divided_by_category[j].append(i)
```

```python
divided_by_category
```

```python
prediction = []
#prediction is a list that stores the category predicted by the knn function from the given test data

for i in X_test.values.tolist():
    #getting and appending the predicted category of all the points in X_test
    prediction.append(knn(divided_by_category, i, 5))

print(prediction)
```

```python
#checking the accuracy of the knn algorithm
count = 0
for i, j in zip(prediction, y_test):
    #if the predicted category matches the actual category from y_list, count is increased 
    if i==j:
        count += 1

print(count/len(y_test.values))
#dividing the number of categories predicted correctly by the total number of predictions, we get the accuracy
```
