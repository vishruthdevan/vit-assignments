```python
# custom knn prediction algorithm

import math


def knn(points, p, k=5):
    distances = [] 
    # list that stores the euclidean distances of all the train data points from the given point p
    
    frequencies = {} 
    # dictionary that stores the number of points of each category neighbouring the given point p 
    
    for category in points:
        frequencies[category] = 0 # initially the number of neighbouring points is 0 
        # calculating the euclidean distance for all features(dimensions) of each point
        for point in points[category]:
            ed = 0  
            for i, j in zip(point, p):
                ed += (i-j)**2
            ed = math.sqrt(ed)
            distances.append((ed, category))
            
    distances = sorted(distances)[:k]
            
    # number of neighbouring points updated for each category
    for i in distances:
        frequencies[i[1]] += 1
    
    # finding the most frequent category in the k neighbouring points
    max = list(frequencies.keys())[0]
    for i in frequencies:
        if frequencies[i] > frequencies[max]:
            max = i
    
    # returning the most frequent category neighbouring the given point
    return max
```

```python
# formatting the data to send as arguments to the knn() function

#converting dataframe values to a list
X_train_list = X_train.values.tolist()
y_train_list = y_train.values.tolist()


divided_by_category = {x:[] for x in y_train.unique()}
# divided_by_category is a dictionary of the form:
# {'B': [list of points with category 'B'], 
#  'M': [list of points with category 'M']}


for i, j in zip(X_train_list, y_train_list):
    # appending the training points to one of the categories in the dictionary
    divided_by_category[j].append(i)
```

```python
# using the knn() function to predict the categories

prediction = []
# prediction is a list that stores the category predicted by the knn() function from the given test data

for i in X_test.values.tolist():
    # calling knn() and appending the predicted category for each point in X_test
    prediction.append(knn(divided_by_category, i, 5))

print(prediction)
```

```python
# checking the accuracy of the knn algorithm

count = 0
for i, j in zip(prediction, y_test):
    # if the predicted category matches the actual category from y_test, count is increased 
    if i==j:
        count += 1

print(count/len(y_test.values))
# dividing the number of categories predicted correctly by the total number of predictions, we get the accuracy
```
